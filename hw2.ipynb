## A 排序

```python
## add your code here
#include <iostream>
#include <vector>
using namespace std;

int main() {
    ios::sync_with_stdio(false);
    cin.tie(nullptr);

    int n, a, b;
    cin >> n >> a >> b;

    vector<int> p(n);
    for (int i = 0; i < n; i++) cin >> p[i];

    int cha1 = (b - a + n) % n;
    int cha2 = (a - b + n) % n;

    vector<int> neng(n, 0);

    for (int d = 1; d < n; d++) {
        for (int w = 0; w < n; w++) {
            int cha = ((w ^ d) - w + n) % n;
            if (cha == cha1 || cha == cha2) {
                neng[d] = 1;
                break;
            }
        }
    }

    for (int m = 0; m < n; m++) {
        int c = (p[0] - m + n) % n;
        vector<int> cuo;

        for (int i = 0; i < n; i++) {
            int now = ((i ^ m) + c) % n;
            if (now != p[i]) {
                cuo.push_back(i);
                if (cuo.size() > 2) break;
            }
        }

        if (cuo.empty()) {
            vector<pair<int, int>> ans;

            if (c != 0) ans.push_back({2, (n - c) % n});
            if (m != 0) ans.push_back({1, m});

            cout << ans.size() << '\n';

            for (auto x : ans) {
                cout << x.first << ' ' << x.second << '\n';
            }

            return 0;
        }

        if (cuo.size() != 2) continue;

        int x = cuo[0], y = cuo[1];
        int fx = ((x ^ m) + c) % n;
        int fy = ((y ^ m) + c) % n;

        if (p[x] != fy || p[y] != fx) continue;

        int d = x ^ y;
        if (!neng[d]) continue;

        int ss = -1, tt = -1;

        for (int s = 0; s < n; s++) {
            int u = p[x] ^ s;
            int v = p[y] ^ s;
            int cha = (v - u + n) % n;

            if (cha == cha1) {
                int t = (a - u + n) % n;
                if ((v + t) % n == b) {
                    ss = s;
                    tt = t;
                    break;
                }
            }
        }

        if (ss == -1) {
            for (int s = 0; s < n; s++) {
                int u = p[x] ^ s;
                int v = p[y] ^ s;
                int cha = (v - u + n) % n;

                if (cha == cha2) {
                    int t = (b - u + n) % n;
                    if ((v + t) % n == a) {
                        ss = s;
                        tt = t;
                        break;
                    }
                }
            }
        }

        if (ss == -1) continue;

        vector<pair<int, int>> ans;

        if (ss != 0) ans.push_back({1, ss});
        if (tt != 0) ans.push_back({2, tt});

        ans.push_back({0, 0});

        if (tt != 0) ans.push_back({2, (n - tt) % n});
        if (ss != 0) ans.push_back({1, ss});

        if (c != 0) ans.push_back({2, (n - c) % n});
        if (m != 0) ans.push_back({1, m});

        cout << ans.size() << '\n';

        for (auto x : ans) {
            if (x.first == 0) cout << 0 << '\n';
            else cout << x.first << ' ' << x.second << '\n';
        }

        return 0;
    }

    cout << -1 << '\n';
    return 0;
}
```

## B 长跑

```python
## add your code here
#include <iostream>
#include <vector>
#include <map>
#include <algorithm>
using namespace std;

int main() {
    ios::sync_with_stdio(false);
    cin.tie(nullptr);

    int n, l, maxn, s;

    while (cin >> n >> l >> maxn >> s) {
        map<int, int> dian;

        for (int i = 0; i < n; i++) {
            int p, c;
            cin >> p >> c;

            if (p > 0 && p < l) {
                if (dian.count(p)) dian[p] = min(dian[p], c);
                else dian[p] = c;
            }
        }

        vector<int> weizhi;
        vector<int> feiyong;

        weizhi.push_back(0);
        feiyong.push_back(0);

        for (auto x : dian) {
            weizhi.push_back(x.first);
            feiyong.push_back(x.second);
        }

        weizhi.push_back(l);
        feiyong.push_back(0);

        int m = weizhi.size();
        int inf = 1000000000;

        vector<int> dp(m, inf);
        dp[0] = 0;

        for (int i = 0; i < m; i++) {
            if (dp[i] == inf) continue;

            for (int j = i + 1; j < m; j++) {
                if (weizhi[j] - weizhi[i] > maxn) break;

                int add = 0;
                if (j != m - 1) add = feiyong[j];

                if (dp[i] + add <= s) {
                    dp[j] = min(dp[j], dp[i] + add);
                }
            }
        }

        if (dp[m - 1] <= s) cout << "Yes\n";
        else cout << "No\n";
    }

    return 0;
}
```

## C 最长回文

```python
## add your code here
#include <iostream>
#include <vector>
#include <string>
#include <algorithm>
using namespace std;

vector<int> guo(const vector<int>& sa, const vector<int>& rk, int pian, int cntt) {
    int n = sa.size();
    vector<int> tong(cntt + 1, 0);
    vector<int> res(n);

    for (int x : sa) {
        int key = 0;
        if (x + pian < n) key = rk[x + pian] + 1;
        tong[key]++;
    }

    for (int i = 1; i <= cntt; i++) {
        tong[i] += tong[i - 1];
    }

    for (int i = n - 1; i >= 0; i--) {
        int x = sa[i];
        int key = 0;
        if (x + pian < n) key = rk[x + pian] + 1;
        res[--tong[key]] = x;
    }

    return res;
}

vector<int> qiu_sa(const string& s) {
    int n = s.size();

    vector<int> sa(n);
    vector<int> rk(n);
    vector<int> nrk(n);
    vector<int> tong(256, 0);

    for (unsigned char ch : s) {
        tong[ch]++;
    }

    for (int i = 1; i < 256; i++) {
        tong[i] += tong[i - 1];
    }

    for (int i = n - 1; i >= 0; i--) {
        sa[--tong[(unsigned char)s[i]]] = i;
    }

    int cntt = 1;
    rk[sa[0]] = 0;

    for (int i = 1; i < n; i++) {
        if (s[sa[i]] != s[sa[i - 1]]) cntt++;
        rk[sa[i]] = cntt - 1;
    }

    for (int k = 1; k < n && cntt < n; k <<= 1) {
        sa = guo(sa, rk, k, cntt);
        sa = guo(sa, rk, 0, cntt);

        nrk[sa[0]] = 0;
        int xin = 1;

        for (int i = 1; i < n; i++) {
            int x = sa[i - 1];
            int y = sa[i];

            int x1 = rk[x];
            int y1 = rk[y];

            int x2 = -1;
            int y2 = -1;

            if (x + k < n) x2 = rk[x + k];
            if (y + k < n) y2 = rk[y + k];

            if (x1 != y1 || x2 != y2) xin++;

            nrk[y] = xin - 1;
        }

        rk.swap(nrk);
        cntt = xin;
    }

    return sa;
}

struct Cha {
    string s;
    int n;
    vector<int> sa;
    vector<int> rk;
    vector<int> h;
    vector<int> lg;
    vector<vector<int>> st;

    Cha(const string& str) {
        s = str;
        n = s.size();

        sa = qiu_sa(s);

        rk.assign(n, 0);

        for (int i = 0; i < n; i++) {
            rk[sa[i]] = i;
        }

        h.assign(n, 0);

        int now = 0;

        for (int i = 0; i < n; i++) {
            int p = rk[i];

            if (p == 0) continue;

            int j = sa[p - 1];

            while (i + now < n && j + now < n && s[i + now] == s[j + now]) {
                now++;
            }

            h[p] = now;

            if (now) now--;
        }

        lg.assign(n + 1, 0);

        for (int i = 2; i <= n; i++) {
            lg[i] = lg[i / 2] + 1;
        }

        int m = lg[n] + 1;

        st.assign(m, vector<int>(n, 0));
        st[0] = h;

        for (int i = 1; i < m; i++) {
            int len = 1 << i;

            for (int j = 0; j + len <= n; j++) {
                st[i][j] = min(st[i - 1][j], st[i - 1][j + (len >> 1)]);
            }
        }
    }

    int wen(int x, int y) {
        if (x == y) return n - x;

        x = rk[x];
        y = rk[y];

        if (x > y) swap(x, y);

        x++;

        int len = y - x + 1;
        int k = lg[len];

        return min(st[k][x], st[k][y - (1 << k) + 1]);
    }
};

void manacher(const string& s, vector<int>& ji, vector<int>& ou) {
    int n = s.size();

    ji.assign(n, 0);
    ou.assign(n, 0);

    int l = 0;
    int r = -1;

    for (int i = 0; i < n; i++) {
        int k = 1;

        if (i <= r) k = min(ji[l + r - i], r - i + 1);

        while (i - k >= 0 && i + k < n && s[i - k] == s[i + k]) {
            k++;
        }

        ji[i] = k;

        if (i + k - 1 > r) {
            l = i - k + 1;
            r = i + k - 1;
        }
    }

    l = 0;
    r = -1;

    for (int i = 0; i < n; i++) {
        int k = 0;

        if (i <= r) k = min(ou[l + r - i + 1], r - i + 1);

        while (i - k - 1 >= 0 && i + k < n && s[i - k - 1] == s[i + k]) {
            k++;
        }

        ou[i] = k;

        if (i + k - 1 > r) {
            l = i - k;
            r = i + k - 1;
        }
    }
}

int suan(int n, const string& a, const string& b) {
    vector<int> jia;
    vector<int> oua;
    vector<int> jib;
    vector<int> oub;

    manacher(a, jia, oua);
    manacher(b, jib, oub);

    string ra = a;
    reverse(ra.begin(), ra.end());

    string all = ra + "{" + b;

    Cha cha(all);

    auto lcp = [&](int x, int y) {
        if (x < 0 || x >= n || y < 0 || y >= n) return 0;

        int res = cha.wen(x, n + 1 + y);
        int lim = min(n - x, n - y);

        if (res > lim) res = lim;

        return res;
    };

    int ans = 0;

    for (int x : jia) ans = max(ans, x * 2 - 1);
    for (int x : oua) ans = max(ans, x * 2);
    for (int x : jib) ans = max(ans, x * 2 - 1);
    for (int x : oub) ans = max(ans, x * 2);

    for (int i = 0; i < n; i++) {
        int r = jib[i];
        int l = i - r + 1;
        int q = i + r;
        int x = lcp(n - 1 - l, q);
        ans = max(ans, r * 2 - 1 + x * 2);
    }

    for (int i = 0; i < n; i++) {
        int r = oub[i];
        int l = i - r;
        int q = i + r;
        int x = lcp(n - 1 - l, q);
        ans = max(ans, r * 2 + x * 2);
    }

    for (int i = 0; i < n; i++) {
        int r = jia[i];
        int l = i - r + 1;
        int rr = i + r - 1;
        int x = lcp(n - l, rr);
        ans = max(ans, r * 2 - 1 + x * 2);
    }

    for (int i = 0; i < n; i++) {
        int r = oua[i];

        if (r == 0) continue;

        int l = i - r;
        int rr = i + r - 1;
        int x = lcp(n - l, rr);
        ans = max(ans, r * 2 + x * 2);
    }

    return ans;
}

int main() {
    ios::sync_with_stdio(false);
    cin.tie(nullptr);

    int n;
    string a;
    string b;

    cin >> n >> a >> b;

    cout << suan(n, a, b) << '\n';

    return 0;
}
```

## D 优惠券

```python
## add your code here
#include <iostream>
#include <vector>
#include <string>
using namespace std;

class Shu {
public:
    int n;
    vector<int> a;

    Shu(int x = 0) {
        chushi(x);
    }

    void chushi(int x) {
        n = x;
        a.assign(n + 2, 0);
    }

    void jia(int x, int v) {
        while (x <= n) {
            a[x] += v;
            x += x & -x;
        }
    }

    int qian(int x) {
        int res = 0;
        while (x > 0) {
            res += a[x];
            x -= x & -x;
        }
        return res;
    }

    int zhao(int k) {
        int now = 0;
        int b = 1;

        while ((b << 1) <= n) b <<= 1;

        while (b) {
            int nxt = now + b;

            if (nxt <= n && a[nxt] < k) {
                now = nxt;
                k -= a[nxt];
            }

            b >>= 1;
        }

        return now + 1;
    }

    int qu(int l) {
        int qianmian = 0;

        if (l > 1) qianmian = qian(l - 1);

        int zong = qian(n);

        if (zong == qianmian) return -1;

        int p = zhao(qianmian + 1);

        jia(p, -1);

        return p;
    }
};

int main() {
    ios::sync_with_stdio(false);
    cin.tie(nullptr);

    int m;

    while (cin >> m) {
        Shu shu(m);
        vector<int> zhuangtai(100005, 0);

        int ans = -1;

        for (int i = 1; i <= m; i++) {
            string op;
            cin >> op;

            int x = 0;

            if (op == "I" || op == "O") cin >> x;

            if (ans != -1) continue;

            if (op != "I" && op != "O") {
                shu.jia(i, 1);
            } else if (op == "I") {
                if (zhuangtai[x] > 0) {
                    if (shu.qu(zhuangtai[x]) == -1) {
                        ans = i;
                        continue;
                    }
                }

                zhuangtai[x] = i;
            } else {
                if (zhuangtai[x] <= 0) {
                    if (shu.qu(-zhuangtai[x]) == -1) {
                        ans = i;
                        continue;
                    }
                }

                zhuangtai[x] = -i;
            }
        }

        cout << ans << '\n';
    }

    return 0;
}
```

## E 任意点

```python
## add your code here
#include <iostream>
#include <vector>
using namespace std;

const int N = 2005;

int fa[N];
int dep[N];

void chushi() {
    for (int i = 0; i < N; i++) {
        fa[i] = i;
        dep[i] = 1;
    }
}

int zhao(int x) {
    if (fa[x] != x) {
        fa[x] = zhao(fa[x]);
    }
    return fa[x];
}

void hebing(int x, int y) {
    x = zhao(x);
    y = zhao(y);

    if (x == y) return;

    if (dep[x] < dep[y]) {
        fa[x] = y;
    } else {
        fa[y] = x;
        if (dep[x] == dep[y]) dep[x]++;
    }
}

int main() {
    ios::sync_with_stdio(false);
    cin.tie(nullptr);

    chushi();

    bool youx[1005] = {};
    bool youy[1005] = {};

    int n;
    cin >> n;

    for (int i = 0; i < n; i++) {
        int x, y;
        cin >> x >> y;

        youx[x] = true;
        youy[y] = true;

        hebing(x, 1000 + y);
    }

    vector<int> dian;

    for (int i = 1; i <= 1000; i++) {
        if (youx[i]) dian.push_back(i);
    }

    for (int i = 1; i <= 1000; i++) {
        if (youy[i]) dian.push_back(1000 + i);
    }

    bool vis[N] = {};
    int ans = 0;

    for (int x : dian) {
        int f = zhao(x);

        if (!vis[f]) {
            vis[f] = true;
            ans++;
        }
    }

    cout << ans - 1 << '\n';

    return 0;
}
```

## F 通配符匹配

```python
## add your code here
#include <iostream>
#include <vector>
#include <string>
#include <algorithm>
using namespace std;

using ull = unsigned long long;

const ull JIN = 1315423911ull;

struct Kuai {
    int pian;
    int len;
    ull h;
};

struct Duan {
    string s;
    int len;
    vector<Kuai> kuai;
    int zhu;
};

vector<ull> mi;

ull quha(const vector<ull>& qian, int l, int len) {
    return qian[l + len] - qian[l] * mi[len];
}

ull suanha(const string& s, int l, int len) {
    ull h = 0;

    for (int i = 0; i < len; i++) {
        h = h * JIN + (s[l + i] - 'a' + 1);
    }

    return h;
}

Duan zuoduan(const string& s) {
    Duan d;
    d.s = s;
    d.len = s.size();
    d.zhu = -1;

    int n = s.size();
    int i = 0;

    while (i < n) {
        if (s[i] == '?') {
            i++;
            continue;
        }

        int j = i;

        while (j < n && s[j] != '?') {
            j++;
        }

        Kuai k;
        k.pian = i;
        k.len = j - i;
        k.h = suanha(s, i, j - i);

        d.kuai.push_back(k);

        i = j;
    }

    for (int i = 0; i < (int)d.kuai.size(); i++) {
        if (d.zhu == -1 || d.kuai[i].len > d.kuai[d.zhu].len) {
            d.zhu = i;
        }
    }

    return d;
}

bool panduan(const string& t, const vector<ull>& qian, const Duan& d, int pos) {
    if (pos < 0 || pos + d.len > (int)t.size()) {
        return false;
    }

    for (auto k : d.kuai) {
        if (quha(qian, pos + k.pian, k.len) != k.h) {
            return false;
        }
    }

    for (auto k : d.kuai) {
        for (int i = 0; i < k.len; i++) {
            if (t[pos + k.pian + i] != d.s[k.pian + i]) {
                return false;
            }
        }
    }

    return true;
}

int zhaoduan(const string& t, const vector<ull>& qian, const Duan& d, int start) {
    int n = t.size();

    if (start + d.len > n) {
        return -1;
    }

    if (d.kuai.empty()) {
        return start;
    }

    Kuai k = d.kuai[d.zhu];
    int r = n - d.len;

    for (int pos = start; pos <= r; pos++) {
        if (quha(qian, pos + k.pian, k.len) == k.h) {
            if (panduan(t, qian, d, pos)) {
                return pos;
            }
        }
    }

    return -1;
}

bool pipei(const string& t, const vector<Duan>& duan, bool kai, bool wei, bool youxing) {
    vector<ull> qian(t.size() + 1, 0);

    for (int i = 0; i < (int)t.size(); i++) {
        qian[i + 1] = qian[i] * JIN + (t[i] - 'a' + 1);
    }

    if (!youxing) {
        if (duan.empty()) {
            return t.empty();
        }

        if ((int)t.size() != duan[0].len) {
            return false;
        }

        return panduan(t, qian, duan[0], 0);
    }

    if (duan.empty()) {
        return true;
    }

    int now = 0;
    int id = 0;
    int last = duan.size() - 1;

    if (!kai) {
        if (!panduan(t, qian, duan[0], 0)) {
            return false;
        }

        now = duan[0].len;
        id = 1;
    }

    int stop = wei ? duan.size() : last;

    while (id < stop) {
        int p = zhaoduan(t, qian, duan[id], now);

        if (p == -1) {
            return false;
        }

        now = p + duan[id].len;
        id++;
    }

    if (!wei) {
        int pos = (int)t.size() - duan[last].len;

        if (pos < now) {
            return false;
        }

        if (!panduan(t, qian, duan[last], pos)) {
            return false;
        }
    }

    return true;
}

int main() {
    ios::sync_with_stdio(false);
    cin.tie(nullptr);

    string p;
    cin >> p;

    int n;
    cin >> n;

    mi.assign(100005, 1);

    for (int i = 1; i < (int)mi.size(); i++) {
        mi[i] = mi[i - 1] * JIN;
    }

    bool kai = p[0] == '*';
    bool wei = p.back() == '*';
    bool youxing = false;

    vector<Duan> duan;
    string now;

    for (char c : p) {
        if (c == '*') {
            youxing = true;

            if (!now.empty()) {
                duan.push_back(zuoduan(now));
                now.clear();
            }
        } else {
            now.push_back(c);
        }
    }

    if (!now.empty()) {
        duan.push_back(zuoduan(now));
    }

    while (n--) {
        string name;
        cin >> name;

        if (pipei(name, duan, kai, wei, youxing)) {
            cout << "YES\n";
        } else {
            cout << "NO\n";
        }
    }

    return 0;
}
```

## G 汉诺塔

```python
## add your code here
#include <iostream>
#include <string>
#include <vector>
using namespace std;

long long mi(long long a, int b) {
    long long ans = 1;

    while (b--) {
        ans *= a;
    }

    return ans;
}

int bianhao(char c) {
    if (c == 'A') return 0;
    if (c == 'B') return 1;
    return 2;
}

int main() {
    ios::sync_with_stdio(false);
    cin.tie(nullptr);

    int n;
    cin >> n;

    vector<string> cao(6);

    for (int i = 0; i < 6; i++) {
        cin >> cao[i];
    }

    vector<int> xia(3, -1);

    for (int i = 0; i < 6; i++) {
        int l = bianhao(cao[i][0]);
        int r = bianhao(cao[i][1]);

        if (xia[l] == -1) {
            xia[l] = r;
        }
    }

    long long ans = 0;

    if (xia[xia[xia[0]]] == 0 && xia[0] != xia[1] && xia[1] != xia[2] && xia[0] != xia[2]) {
        ans = mi(2, n) - 1;
    } else if (xia[1] == 2 && xia[2] == 1) {
        ans = mi(3, n - 1);
    } else {
        ans = 2 * mi(3, n - 1) - 1;
    }

    cout << ans << '\n';

    return 0;
}
```

## H 马步距离

```python
## add your code here
#include <iostream>
#include <algorithm>
using namespace std;

int main() {
    ios::sync_with_stdio(false);
    cin.tie(nullptr);

    long long x1, y1, x2, y2;
    cin >> x1 >> y1 >> x2 >> y2;

    long long x = x1 - x2;
    long long y = y1 - y2;

    if (x < 0) x = -x;
    if (y < 0) y = -y;

    if (x < y) swap(x, y);

    long long ans = 0;

    if (x == 0 && y == 0) {
        ans = 0;
    } else if (x == 1 && y == 0) {
        ans = 3;
    } else if (x == 2 && y == 2) {
        ans = 4;
    } else {
        long long a = (x + 1) / 2;
        long long b = (x + y + 2) / 3;

        ans = max(a, b);

        if ((ans + x + y) % 2 != 0) {
            ans++;
        }
    }

    cout << ans << '\n';

    return 0;
}
```

## I 直方图最大矩形

```python
## add your code here
class Solution {
public:
    /**
     * 代码中的类名、方法名、参数名已经指定，请勿修改，直接返回方法规定的值即可
     *
     * 
     * @param heights int整型vector 
     * @return int整型
     */
    int largestRectangleArea(vector<int>& heights) {
        int n = heights.size();
        vector<int> zhan;
        int ans = 0;

        for (int i = 0; i <= n; i++) {
            int now;

            if (i == n) now = 0;
            else now = heights[i];

            while (!zhan.empty() && now < heights[zhan.back()]) {
                int gao = heights[zhan.back()];
                zhan.pop_back();

                int zuo;

                if (zhan.empty()) zuo = -1;
                else zuo = zhan.back();

                int kuan = i - zuo - 1;
                ans = max(ans, gao * kuan);
            }

            zhan.push_back(i);
        }

        return ans;
    }
};
```

## J 消防局的设立

```python
## add your code here
#include <iostream>
#include <vector>
#include <algorithm>
using namespace std;

int main() {
    ios::sync_with_stdio(false);
    cin.tie(nullptr);

    int n;
    cin >> n;

    vector<int> fu(n + 1, 0);
    vector<int> shen(n + 1, 0);
    vector<int> xu;

    for (int i = 1; i <= n; i++) {
        xu.push_back(i);
    }

    for (int i = 2; i <= n; i++) {
        cin >> fu[i];
        shen[i] = shen[fu[i]] + 1;
    }

    sort(xu.begin(), xu.end(), [&](int a, int b) {
        return shen[a] > shen[b];
    });

    vector<int> huo(n + 1, 0);
    vector<int> zi(n + 1, 0);
    vector<int> sun(n + 1, 0);

    int ans = 0;

    for (int u : xu) {
        int f = fu[u];
        int ff = 0;

        if (f) ff = fu[f];

        bool ok = false;

        if (huo[u]) ok = true;
        if (f && huo[f]) ok = true;
        if (ff && huo[ff]) ok = true;
        if (zi[u]) ok = true;
        if (sun[u]) ok = true;
        if (f && zi[f]) ok = true;

        if (!ok) {
            int p = u;

            if (fu[p]) p = fu[p];
            if (fu[p]) p = fu[p];

            huo[p] = 1;
            ans++;

            int fp = fu[p];
            int ffp = 0;

            if (fp) ffp = fu[fp];

            if (fp) zi[fp]++;
            if (ffp) sun[ffp]++;
        }
    }

    cout << ans << '\n';

    return 0;
}
```
